 2. Analizar la calidad y confiabilidad de la información registrada por los sensores.

2.1. Timestamps y frecuencia de muestreo
	1.	Fechas futuras / incorrectas
	•	10 registros con fecha 2030-05-15 00:00:00 (todas las filas con ese timestamp).
	•	Esto sugiere un problema de RTC/fecha en el dispositivo o en el servidor.
	2.	Registros duplicados en clave (timestamp + device_id)
	•	Filas completamente duplicadas: 8
	•	Duplicados por pareja (timestamp, device_id): 16 pares
→ En total, 29 filas involucradas en duplicados (≈ 11.2% del dataset).
	•	En un par de casos, mismo timestamp y medición pero distinto status (OK vs FAIL), lo que revela inconsistencia en el estado reportado.
	3.	Frecuencia de muestreo irregular
	•	Ideal: cada 30 minutos.
	•	En la práctica, diferencias entre lecturas de 30, 60, 90, 120 min o más.
	•	Cobertura aproximada (solo periodo 2023, sin limpiar):
	•	ESP32_A1: ~35% de los slots esperados.
	•	ESP32_A2: ~33%.
	•	ESP32_B1: ~30%.

Conclusión: hay lagunas de datos y registros repetidos, lo que indica fallos de transmisión/almacenamiento además del error de fecha.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- Cargar datos ---
file_path = r"C:\Users\AMB-CTEC-CCSIM-01\Downloads\iot_sensor_corrupted_dataset.csv"
df = pd.read_csv(file_path)

# --- Preprocesamiento ---
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")

# 1. Timestamps incorrectos (año 2030)
errores_timestamp = sum(df["timestamp"].dt.year > 2023)

# 2. Filas duplicadas completas
duplicados_completos = df.duplicated().sum()

# 3. Duplicados por clave (timestamp + device_id)
duplicados_clave = df.duplicated(subset=["timestamp", "device_id"]).sum()

# 4. Frecuencia de muestreo irregular (GRAN visión resumida)
# Calcular diferencias entre timestamps por dispositivo
def irregularidad(dev):
    sub = df[df["device_id"] == dev].sort_values("timestamp")
    diffs = sub["timestamp"].diff().dt.total_seconds() / 60  # minutos
    # Consideramos irregular todo lo que NO sea 30 minutos exactos
    return sum((diffs != 30) & (~diffs.isna()))

irreg_A1 = irregularidad("ESP32_A1")
irreg_A2 = irregularidad("ESP32_A2")
irreg_B1 = irregularidad("ESP32_B1")

# empaquetar valores
metricas = {
    "Timestamps incorrectos (2030)": errores_timestamp,
    "Filas duplicadas completas": duplicados_completos,
    "Duplicados por timestamp+device_id": duplicados_clave,
    "Muestreo irregular ESP32_A1": irreg_A1,
    "Muestreo irregular ESP32_A2": irreg_A2,
    "Muestreo irregular ESP32_B1": irreg_B1
}

# --- Gráfica ---
plt.figure(figsize=(12,6))
plt.bar(metricas.keys(), metricas.values())
plt.title("Indicadores de Calidad y Confiabilidad del Dataset IoT")
plt.ylabel("Cantidad de casos detectados")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```

6. Conclusiones técnicas y recomendaciones para mantenimiento

6.1. ¿Qué datos considero válidos?

Para toma de decisiones operativas yo usaría:
	1.	Solo registros con:
	•	timestamp entre 2023-01-01 y 2023-01-06,
	•	device_id en {ESP32_A1, ESP32_A2, ESP32_B1},
	•	Temperatura [15, 40] °C,
	•	Humedad [0, 100]%,
	•	Sin duplicados en (device_id, timestamp).
	2.	Esto deja 193 registros (~74%) que:
	•	Parecen físicamente coherentes,
	•	Presentan valores estables y razonables,
	•	Permiten estimar condiciones promedio por área/laboratorio.

Los demás datos deben ser descartados o corregidos (si existiera una fuente externa para reconstruirlos, que aquí no la tenemos).


6.2. Recomendaciones para el área de mantenimiento / TI

A nivel de hardware/sensores:
	1.	Revisar ESP32_A1:
	•	Es el sensor con más fallas y anomalías.
	•	Posibles acciones: recalibración, revisión de conexiones, prueba de reemplazo temporal del módulo.
	2.	Verificar la alimentación y conectividad de todos los ESP32:
	•	Gaps de 60–150 min sugieren cortes de energía, problemas de WiFi o reinicios frecuentes.
	3.	Revisión del reloj (RTC/NTP):
	•	Las lecturas de 2030-05-15 indican que el reloj del dispositivo o del servidor se desconfigura.
	•	Asegurar sincronización NTP y manejo robusto de hora en firmware.


A nivel de firmware / software en los ESP32:
	1.	Evitar valores centinela como 999 °C o -10% sin marcar fallo:
	•	En vez de “inventar” 999 o -10, enviar explícitamente:
	•	status = FAIL
	•	temperature = null / humidity = null
	•	Así se evita mezclar códigos de error con lecturas numéricas.
	2.	Normalizar el estado:
	•	Limitar status a dos valores exactos: OK y FAIL.
	•	Evitar variantes (ok, Ok, FAIL ).
	3.	Validaciones en el dispositivo:
	•	No enviar lecturas fuera de rangos físicos razonables (por ejemplo, temp < -10 o > 60 °C, humedad < 0 o > 100%).


A nivel de backend / servidor de datos:
	1.	Aplicar reglas de calidad al momento de insertar datos:
	•	Rechazar o marcar con flag:
	•	Fechas futuras o demasiado antiguas.
	•	Valores fuera de rango.
	•	device_id desconocidos o vacíos.
	•	Campos no numéricos ('??').
	2.	Deduplicación en la capa de ingestión:
	•	Definir (device_id, timestamp) como clave única.
	•	Si llega más de un registro para la misma clave:
	•	Mantener uno solo (por ejemplo, el último) o
	•	Registrar ambos pero marcando uno como duplicado.
	3.	Monitoreo de cobertura:
	•	Implementar panel que muestre % de slots de 30 minutos llenos por día y por dispositivo.
	•	Si la cobertura baja de cierto umbral (por ejemplo, 90%), disparar alerta para mantenimiento.
